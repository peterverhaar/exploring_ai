# Definitions and terminology

Artificial Intelligence has been defined as "making a machine behave in ways that would be called intelligent if a human were so behaving"
John McCarthy et al, “A Proposal for the Dartmouth Summer Research Project on Artificial Intelligence”, 1955. https://doi.org/10.1609/aimag.v27i4.1904


There are many different forms of AI:

* Generative AI include forms of AI that can generate text, images, and other content based on data sets they were trained on.
* Descriptive AI can used to classify text, images, audio recording and video files. It is typically used to generate labels that can help to enhance the access to these reources. Computer Vision is a specific form of descriptive AI that can be used, among other things, to classify objects in images and in video files.


## Methods used to train AI models

A machine learning algorithm is designed to recognize patterns, relationships, and trends in data and will adjust  the internal parameters based on examples to improve their performance over time. It is the capability of a computer to learn a specific actitivy without being explicitly programmed. The computer mainly learns from examples it is given.
Deep Learning is a specific type of machine learning which is based on Neural Networks. [elaborate]

## Large Language Models

A large language model (LLM) is a type of machine learning model that can perform natural language processing tasks such as generating and classifying text, answering questions, and translating text.
Language models are a type of machine learning model designed to predict the likelihood of a sequence of text, which means that they can be set up to predict the most likely way to continue a conversation. The models behind ChatGPT are large, complex neural networks that have been exposed to an enormous amount of text from books, articles, websites, and more. These models have learned something about patterns in grammar and word meaning, including the way that meaning arises contextually across multiple sentences and multiple turns in a conversation. When you ask ChatGPT a question, you are presenting the model with new information it tries to make a prediction on, in this case, it tries to generate a response that matches the pattern of conversation. Large language models are trained on massive amounts of text data and are essentially the backbone of NLP today. There are and have been for many years, large language models out there actually but ChatGPT has currently caught the popular imagination because of its publicly available interface and remarkable performance.

```python
from transformers import pipeline
import requests
import os
import PyPDF2
import re
import transformers
import nltk
from spacy.lang.en import English
from transformers import AutoTokenizer, AutoModelForTokenClassification
```

## Download a PDF file and extract all the text

```python
url = 'https://barcelona-declaration.org/downloads/BarcelonaDeclaration.pdf'
file_name = os.path.basename(url)

with open(file_name, "wb") as file:
    response = requests.get(url)
    file.write(response.content)

full_text = ''

pdf_file = open(file_name, 'rb')
read_pdf = PyPDF2.PdfReader(pdf_file)
for page in read_pdf.pages:
    full_text += page.extract_text() + ' '

full_text = re.sub( r'\s+' , ' ' , full_text )
print(full_text)
```

The full text of the Barcelon Declaration is too long. We need to divide the full text into chunks of 1000 words.

```python
chunks = []

nlp = English()
nlp.add_pipe('sentencizer')
doc = nlp(full_text)
sentences = [sent.text.strip() for sent in doc.sents]

chunk_length = 0
chunk_text = ''

for sentence in sentences:
  words = re.split('\s+',sentence)
  chunk_length += len(words)
  if chunk_length < 1000:
    chunk_text += sentence
  else:
    chunks.append(chunk_text)
    chunk_text = sentence
    chunk_length = 0


## Summary per chunk

summarizer = pipeline("summarization" , model = 'sshleifer/distilbart-cnn-12-6' )

max_length = 500
min_length = 150


```

# Named Entity recognition

See https://huggingface.co/dslim/bert-base-NER?text=My+name+is+Sarah+and+I+live+in+London

```python
tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")

model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")

nlp = pipeline("ner", model=model, tokenizer=tokenizer)
example = "My name is Wolfgang and I live in Berlin"

ner_results = nlp(example)

```

```python
for i in range(0,len(ner_results)):

  print(f"{ner_results[i]['word']} => {ner_results[i]['entity']}")
```

```python
import requests
import timm
import cv2
import torch
from transformers import DetrImageProcessor, DetrForObjectDetection


#Load the image using OpenCV
image_path = 'image.jpg'
im = cv2.imread(image_path)

# Check if the image was loaded successfully
if im is None:
    raise ValueError(f"Image at path '{image_path}' could not be loaded.")

#Convert the image from BGR (OpenCV default) to RGB
image = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)


#Initialize the image processor and model
processor = DetrImageProcessor.from_pretrained("facebook/detr-resnet-50")
model = DetrForObjectDetection.from_pretrained("facebook/detr-resnet-50")

# Prepare the image for the model
inputs = processor(images=image, return_tensors="pt")

# Perform inference without computing gradients
with torch.no_grad():
    outputs = model(**inputs)

# Process the outputs
target_sizes = torch.tensor([image.shape[:2]])
results = processor.post_process_object_detection(outputs, target_sizes=target_sizes, threshold=0.9)[0]

# Draw the bounding boxes on the image and print the labels
for score, label, box in zip(results["scores"], results["labels"], results["boxes"]):
    box = [round(i, 2) for i in box.tolist()]
    cv2.rectangle(im, (int(box[0]), int(box[1])), (int(box[2]), int(box[3])), (0, 255, 0), 2)
    label_str = f"{model.config.id2label[label.item()]}: {round(score.item(), 3)}"
    cv2.putText(im, label_str, (int(box[0]), int(box[1] - 10)), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    print(label_str)

# Save and display the result image
#cv2.imwrite("result_images.jpg", im)
#cv2_imshow(im)
```
